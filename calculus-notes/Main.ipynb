#### Arc Length Formulas
To find the length of a smooth curve in the coordinate plane, we can use one of the following: \
$S = \int_{a}^{b}\sqrt{1+[f'(x)]^2} \mathrm{d}x$ \
$S = \int_{c}^{d}\sqrt{1+[g'(y)]^2} \mathrm{d}y$

#### Find the length of $f(x) = 3x^3 - 9x + 5$ over the interval [1,6]

```python
from sympy import *
import numpy as np
from scipy.integrate import quad
x = symbols('x')
f = (1/6)*x**3 + (1/2)*(1/x)
deriv = f.diff(x)
integrand = lambdify(x, sqrt(1 + deriv**2), 'numpy')
quad(integrand, 0.5, 2)
```

#### Pappus's Theorem
Pappus's Theorem is used to determine the surface area of a solid generated by rotating a functio around a given axis. The surface is given by: \
$\int_{a}^{b}2\pi R\sqrt{1+(f'(x))^2} \mathrm{d}x$ \
$\int_{a}^{b}2\pi R\sqrt{1+(g'(y))^2 }\mathrm{d}y$ \
where R is the radius of the solid of revolution.

#### The curve $x = \frac{y^4}{4} + \frac{1}{8y^2}, 1 \le y \le 2$, is rotated about the y-axis. Find the surface area of the surface generated.

$2\pi \cdot \int_{1}^{2}(\frac{y^4}{4}+\frac{1}{8y^2})\sqrt{1+(y^3 - \cfrac{1}{4y^3})^2}\mathrm{d}y$ \
$\approx 51.882$

#### Evaluate the area of the surface generated by revolving the curve $y = \frac{x^3}{3}+\frac{1}{4x}, 1 \le x \le 3$, about the line $y = -2$.

$2\pi \cdot \int_{1}^{3}(\frac{x^3}{3}+\frac{1}{4x}+2)\sqrt{1+(x^2 - \cfrac{1}{4x^2})^2} \mathrm{d}x$ \
$\approx 373.675$

#### Find the surface area of the solid genrerated by revolving the region bounded by $f(x) = x^3$ and $gt(x) = -x^2 + 2$ in the first quadrant around the x-axis.

$2\pi \cdot \int_{0}^{1}((-x^2+2))\sqrt{1+(-2x)^2} \mathrm{d}x$ + $2\pi \cdot \int_{0}^{1}x^3\sqrt{1+(3x^2)^2} \mathrm{d}x + 4\pi$ \
$\approx 30.905$

```python
import numpy as np
import matplotlib.pyplot as plt
def three_d_plotter(angle, rotate, turn):
    fig = plt.figure(figsize = (13, 6))
    ax = fig.add_subplot(1, 1, 1, projection='3d')

    u = np.linspace(-1, 2, 60)
    v = np.linspace(0, angle, 60)
    U, V = np.meshgrid(u, v)

    X = U
    Y1 = (U**3)*np.cos(V)
    Z1 = (U**3)*np.sin(V)

    Y2 = (-U**2 + -2)*np.cos(V)
    Z2 = (-U**2 -2)*np.sin(V)

    ax.plot_surface(X, Y1, Z1, alpha=0.3, color='red', rstride=6, cstride=12)
    ax.plot_surface(X, Y2, Z2, alpha=0.3, color='blue', rstride=6, cstride=12)
    ax.plot_wireframe(X, Y2, Z2, alpha=0.3, color='blue', rstride=6, cstride=12)
    ax._axis3don = False
    ax.view_init(elev = rotate, azim = turn)

    plt.show()

from ipywidgets import interact
import ipywidgets as widgets

interact(three_d_plotter, angle = widgets.FloatSlider(0, min = 0, max = 2*np.pi, step = np.pi/10),
        rotate = widgets.FloatSlider(0, min = 0, max = 360, step = 5),
         turn = widgets.FloatSlider(0, min = 0, max = 500, step = 5))
```

### Cross Products (02/21/2023)

The cross product (vector product) of two vecotrs $\vec{v}$ and $\vec{w}$ gives a third vector that is orthogonal to both $\vec{v}$ and $\vec{w}$.

#### Calculating Cross Products
The cross product of three-dimensional vectors can be calculated by creating a 3x3 matrix and finding its determinant.

#### Example: Let $\vec{v} = <a, b, c>$ and $\vec{w} = <x, y, z>$. Then, $\vec{v} \times \vec{w} =$
$\vec{v} \times \vec{w} =$
$\left \lvert \begin{matrix}
\vec{i} & \vec{j} & \vec{k} \\
a & b & c \\
x & y & z 
\end{matrix} \right \rvert
$ 
$ = (b_z - c_y)\hat{i} - (a_z - c_x)\hat{j} + (a_y - b_x)\hat{k}$

#### If $\vec{v} = <-1, 4, 7>$ and $\vec{w} = <2, 5, -3>$, find a vector that is orthogonal to both $\vec{v}$ and $\vec{w}$.

#### Find the cross product of $3i+2j-k$ and $i-9j+5k$

$\left \lvert \begin{matrix}
\vec{i} & \vec{j} & \vec{k} \\
3 & 2 & -1 \\
1 & -9 & 5 
\end{matrix} \right \rvert$ \
$ = \left \lvert \begin{matrix}
2 & -1 \\
-9 & 5 
\end{matrix} \right \rvert i$
$ - \left \lvert \begin{matrix}
3 & -1 \\
1 & 5 
\end{matrix} \right \rvert j$
$ + \left \lvert \begin{matrix}
3 & 2 \\
1 & -9
\end{matrix} \right \rvert k$ \
$=(2 \cdot 5 - -1 \cdot -9)i - (3 \cdot 5 - -1 \cdot 1)j + (3 \cdot -9 - 2 \cdot 1)k$ \
$=(10-9)i - (15+1)j + (-27-2)k$ \
$=i-16j-29k$

#### Find the cross product of $2i - j + k$ and $4i - 2j + 2k$

$\left \lvert \begin{matrix}
\vec{i} & \vec{j} & \vec{k} \\
2 & -1 & 1 \\
4 & -2 & 2 
\end{matrix} \right \rvert$ \
$ = \left \lvert \begin{matrix}
-1 & 1 \\
-2 & 2 
\end{matrix} \right \rvert i$
$ - \left \lvert \begin{matrix}
2 & 1 \\
4 & 2 
\end{matrix} \right \rvert j$
$ + \left \lvert \begin{matrix}
4 & -2
\end{matrix} \right \rvert k$ \
$=(-1 \cdot 2 - 1 \cdot -2)i - (2 \cdot 2 - 1 \cdot 4)j + (2 \cdot -2 - -1 \cdot 4)k$ \
$=(-2+2)i - (4-4)j + (-4+4)k$ \
$=0i-0j+0k$ \
Note: Since the cross product is the zero vector, the two vectors are parallel to each other.

#### Application: Area of a parallelogram
The magnitude of the cross product of two vectors will give the area of a parallelogram defined by the vectors.
